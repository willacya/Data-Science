```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('Pokemon.csv')
```

```python
df.head()
```

```python
df.drop(['Generation','Legendary'], axis=1, inplace=True)
```

```python
sns.jointplot(x='HP', y='Attack', data=df)
```

```python
sns.boxplot(y='HP', data=df)
```

```python
sns.boxplot(data=df)
```

```python
df.drop(['Total','#'],1, inplace=True)
```

```python
sns.boxplot(data=df)
```

```python
df = pd.melt(df, id_vars=['Name','Type 1','Type 2'], var_name='Stat')
```

```python
df.head()
```

```python
sns.swarmplot(x='Stat', y='value',data=df)
```

```python
plt.figure(figsize=(12,10))
plt.ylim(0,270)
sns.swarmplot(x="Stat", y="value", data=df, hue='Type 1', split=False, size=7)
plt.legend(bbox_to_anchor=(1,1), loc=2, borderaxespad=0.)
```

```python
plt.figure(figsize=(12,10))
plt.ylim(0,270)
sns.swarmplot(x="Stat", y="value", data=df, hue='Type 1', split=True, size=7)
plt.legend(bbox_to_anchor=(1,1), loc=2, borderaxespad=0.)
```
